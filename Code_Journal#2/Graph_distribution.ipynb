```python
import matplotlib.pyplot as plt
import numpy as np
```

Select your favorite distribution (I picked gaussian) and pull 1000 random numbers from that distribution.

```python
gauss_numbers = np.random.normal(loc = 100, scale= 2, size = 1000)
```

    Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF.

```python
plt.hist(gauss_numbers, bins=100)
plt.xlabel('gauss_numbers')
plt.ylabel('count')
plt.savefig('gauss_numbers.pdf')
plt.show()
```
