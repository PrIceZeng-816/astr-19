Create a Jupyter notebook, import matplotlib. Write cells that create an array x ranging from [0,1] in 100 steps and that defines a function that returns exp(x).  In a new cell use the function to set y=exp(x), and then plot x vs. y. Label the x-axis as “Time [milliseconds]” and the y-axis as “Awesomeness”. Save the figure as a PDF.

Importing matplotlib

```python
import matplotlib
from matplotlib import pyplot as plt
```

Importing numpy for linspace function

```python
import numpy as np
```

create an array x ranging from [0,1] in 100 steps and that defines a function that returns exp(x).

```python
x = {}
steps = 100
start = 0
end = 1

x = np.linspace(start, end, steps)

def exp(x):
    return np.exp(x)
```

 set y=exp(x), and then plot x vs. y

```python
y = exp(x)
plt.xlabel("Time [millisecond")
plt.ylabel("Awesomeness")
plt.plot(x, y)
plt.savefig('awesome.pdf')
```
